```python
from pathlib import Path
from src.geoplot import GEOPlot
import pandas as pd
import plotly.graph_objects as go
import re
from consolidation import consolidation_settlement
```

```python
key = '200kPa'
OCR = 1
Cc = 1.2
e0 = 2

output_path = Path('./output')
files = output_path.glob(f'*{key}*.json')
fig = GEOPlot.get_figure()
markers = GEOPlot.marker_list(open_marker=True)
ymax=0
for file in files:
    marker = next(markers) 
    df = pd.read_json(file)
    model_name = file.stem.split('-')[0]
    nums = re.findall('\d?\.?\d+',file.stem)
    fig.add_trace(go.Scatter(x = df.time,y= df.y, mode='markers+lines',
                             marker=dict(symbol=marker,size=5, color='black'),
                            line=dict(width=1),name=model_name))
    ymax = df.y.max() if ymax<df.y.max() else ymax
    
fig.update_layout(width=1000, height=600,legend=dict(orientation='h'))
fig.update_yaxes(range=[ymax*1.05,-0.1],title='Settlement(m)',showgrid=True,gridwidth=1, gridcolor='grey')
fig.update_xaxes(title='Time(days)',type='log',showgrid=True,gridwidth=1, gridcolor='grey')
# fig.write_html(f'../image/{file.stem.split('-')[1]}-{file.stem.split('-')}[2].html')'
# fig.write_html(f'../image/{model_name}_H_{nums[0]}.html')
fig.write_image(f'../image/{model_name}_H_{nums[0]}_{key}_OCR={OCR}.pdf')
fig
```

# Let's plot the degree of freedom below

```python
# key = '200kPa'
OCR = 1
Cc = 1.2
e0 = 2

output_path = Path('./output/OCR =1')
# files = output_path.glob(f'*{key}*.json')
files = output_path.glob(f'*.json')
fig = GEOPlot.get_figure()
markers = GEOPlot.marker_list(open_marker=True)
ymax=0
colors = {'20': 'blue','200':'black'}
for file in files:
    marker = next(markers) 
    df = pd.read_json(file)
    model_name = file.stem.split('-')[0]
    if model_name =='Soft_Soil_Creep':
        continue
    nums = re.findall('\d?\.?\d+',file.stem)
    pressure = nums[-1]
    fig.add_trace(go.Scatter(x = df.time,y= df.y/df.y.max(), mode='markers+lines',
                             marker=dict(symbol=marker,size=5, color=colors[pressure]),
                            line=dict(width=1),name=f'{model_name}-{pressure}kPa'))
    ymax = df.y.max() if ymax<df.y.max() else ymax
    
fig.update_layout(width=1000, height=600,legend=dict(orientation='h'))
fig.update_yaxes(range=[1,-0.1],title='Degree of Consolidation (DoC)',showgrid=True,gridwidth=1, gridcolor='grey')
fig.update_xaxes(title='Time(days)',type='log',showgrid=True,gridwidth=1, gridcolor='grey')
# fig.write_html(f'../image/{file.stem.split('-')[1]}-{file.stem.split('-')}[2].html')'
# fig.write_html(f'../image/{model_name}_H_{nums[0]}.html')
fig.write_image(f'../image/{model_name}_H_{nums[0]}-DoC constant k.pdf')
fig
```
